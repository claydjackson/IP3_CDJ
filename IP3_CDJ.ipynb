## #5

```python
car_sales['Price']=car_sales['Price'].str.replace(r'\$|,', '', regex=True)
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
car_sales.head()
```

## #6

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales.head(3)
```

## #7

```python
from datetime import datetime, timedelta
```

```python
My_today=datetime.today().date()
```

```python
car_sales['Sale Date']=pd.date_range(start=My_today, periods=len(car_sales))
```

```python
car_sales.head()
```

## #8

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.title("Prive vs total sales line plot")
plt.show()
```

## #9

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price scatter plot')
plt.show()
```

## #10

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.ylabel('Odometer (KM)')
plt.xlabel('Make')
plt.title('Odometer vs Make scatter plot')
plt.show()
```

## #11

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of Car Make')
plt.show()
```

## #12

```python
fig=plt.figure(figsize=(20,10))
ax1=fig.add_subplot(221) # 2 row, 3 columns, first plot
ax1.set_title("Prive vs total sales line plot")
ax1.plot(car_sales['Price'], car_sales['Total Sales'])
plt.xlabel("Price")
plt.ylabel("Total Sales")

ax2=fig.add_subplot(222) # 2 row, 3 columns, first plot
ax2.set_title("Odometer vs Price scatter plot")
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')

ax3=fig.add_subplot(223) # 2 row, 3 columns, first plot
ax3.set_title("Make vs Odometer bar plot")
ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.ylabel('Odometer (KM)')
plt.xlabel('Make')

ax4=fig.add_subplot(224) # 2 row, 3 columns, first plot
ax4.set_title("Histogram of car make")
ax4.hist(car_sales['Make'])
plt.xlabel('Make')

plt.show
```
